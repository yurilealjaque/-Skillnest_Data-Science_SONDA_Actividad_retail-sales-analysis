## Carga de Datos

Se utiliza `np.genfromtxt()` para cargar el archivo CSV.  
Como el dataset contiene columnas numéricas y de texto, se usa `dtype=None` para que NumPy detecte automáticamente los tipos de datos.

```python
def cargar_datos(ruta_archivo):
    
    datos = np.genfromtxt(ruta_archivo,delimiter=",", names=True, dtype=None, encoding="utf-8")
    
    return datos
```

```python
ruta_archivo = "retail_sales_dataset.csv"

datos = cargar_datos(ruta_archivo)

print(datos[:5]) #se puso 5 para evitar mostrar todos los datos del archivo(modo de ejemplo)
```

```python
ruta_archivo = "retail_sales_dataset.csv"

datos = cargar_datos(ruta_archivo)

print(datos[:5])
```

```python
print(datos.dtype.names)
```

```python
print("Cantidad de registros:", len(datos))
print("Columnas del dataset:")
print(datos.dtype.names)

print("Primeras 5 filas:")
print(datos[:5])
```

```python
categorias = datos["Product_Category"]
cantidades = datos["Quantity"]
precios_unitarios = datos["Price_per_Unit"]
ventas_totales = datos["Total_Amount"]
```

# Ver categorías únicas

```python
categorias_unicas = np.unique(categorias)

print("Categorías de productos:")
print(categorias_unicas)
```

# Calcular total de ventas por categoría


```python
totales_por_categoria = []

for categoria in categorias_unicas:
    filtro = categorias == categoria
    total_categoria = np.sum(ventas_totales[filtro])
    totales_por_categoria.append(total_categoria)

totales_por_categoria = np.array(totales_por_categoria)

print("Categorías únicas:")
print(categorias_unicas)

print("Totales por categoría:")
print(totales_por_categoria)
```

# Promedio general de ventas

```python
print("PROMEDIO DE VENTAS POR CATEGORÍA")

promedios_por_categoria = []

for categoria in categorias_unicas:
    filtro = categorias == categoria
    promedio_categoria = np.mean(ventas_totales[filtro])
    promedios_por_categoria.append(promedio_categoria)
    
    print(f"{categoria}: ${promedio_categoria:.2f}")

promedios_por_categoria = np.array(promedios_por_categoria)
```

# Venta máxima

```python
indice_mayor = np.argmax(totales_por_categoria)

categoria_mayor = categorias_unicas[indice_mayor]
venta_mayor = totales_por_categoria[indice_mayor]

print("Categoría con mayor venta:")
print(f"{categoria_mayor}: ${venta_mayor}")
```

# Venta minima

```python
indice_menor = np.argmin(totales_por_categoria)

categoria_menor = categorias_unicas[indice_menor]
venta_menor = totales_por_categoria[indice_menor]

print("Categoría con menor venta:")
print(f"{categoria_menor}: ${venta_menor}")
```

# Preprocesamiento básico

```python
suma_total_ventas = np.sum(ventas_totales)
promedio_general_ventas = np.mean(ventas_totales)
venta_maxima = np.max(ventas_totales)
venta_minima = np.min(ventas_totales)
total_productos_vendidos = np.sum(cantidades)

print(f"Suma total de ventas: ${suma_total_ventas}")
print(f"Promedio general de ventas: ${promedio_general_ventas:.2f}")
print(f"Venta máxima registrada: ${venta_maxima}")
print(f"Venta mínima registrada: ${venta_minima}")
print(f"Total de productos vendidos: {total_productos_vendidos}")
```

```python
total_calculado = cantidades * precios_unitarios

print("Total calculado con Quantity x Price_per_Unit:")
print(total_calculado[:10])
```

```python
diferencia = ventas_totales - total_calculado

print("Diferencia entre Total_Amount y total calculado:")
print(diferencia[:10])
```

```python
ingreso_promedio_por_unidad = suma_total_ventas / total_productos_vendidos

print(f"Ingreso promedio por unidad vendida: ${ingreso_promedio_por_unidad:.2f}")
```
